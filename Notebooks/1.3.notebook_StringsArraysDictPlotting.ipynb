```julia
#Automatically fill a dictionary based on an array of values

Purse = Dict()
money = [1000, 500, 250, 125, 75]
for (i, p) in enumerate(money)
    Purse["rank$i"] = p
end
Purse   
```

```julia
# a more useful phonebook
struct person
    lastName::String
    age::Int32
    phone::String
end
chris = person("Stricker",56,"+41796433408")
rohan = person("Fernando", 66, "+1(515) 441-2257‬")
chris.age
```

```julia
myPhoneBook = Dict("chris"=>chris, "rohan"=>rohan)
```

```julia
for (key, value) in myPhoneBook
    println("$key has lastname $(value.lastName), 
            is $(value.age) years of age and can 
            be contacted at $(value.phone)")
end
```
